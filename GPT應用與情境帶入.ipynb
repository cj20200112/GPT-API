## 創建completion(可與AI聊天)
### 架構介紹 :
* model : 選擇的模型
* prompt : 輸入的指令
* max_tokens : 產生的token上限
* temperature : 產生文本的隨機性可選0-2之間，越高越隨機
* top_p : 依照詞語的累積機率，產生下一個使用詞語，可選0-1之間越高越隨機
* n : 生成的結果數量
* frequency_penalty : 可選擇-2到2之間，越高代表越抑制文本重複
* echo : 若為True則返回含提示詞(prompt)的完整文本，False只返回生成的部分
* stop : 若遇到某詞時就停止生成

```python
res = client.completions.create(
    model="gpt-3.5-turbo-instruct",
    prompt="列出五種貓咪品種以及一句話介紹",
    max_tokens=200,
    #temperature=1.8,
    #top_p=0.9,
    #n=3,
    #echo=True,
    #stop="貓",
    frequency_penalty=2
    )
```

```python
res.choices  # 生成的元素是字典型式，這邊取出了全部結果
```

```python
print(res.choices[0].text)  # 取出第一次生成結果的內文
```

# 客服機器人

```python
!pip install openai
!pip install python-dotenv
!pip install gradio==3.48.0
import gradio as gr
from openai import OpenAI
from dotenv import dotenv_values
config = dotenv_values('.env')
client = OpenAI(api_key=config["APIKEY"])
```

可自行決定是否提供以下模擬情境給AI進行角色模擬

```python
system = """
您是好好吃餐酒館的老闆，請根據以下資訊，並以和善好客的態度回覆顧客問題

好好吃餐酒館是一家異國料理餐廳,提供各式主食、酒類飲品、與高級桌邊服務的同時，也聘任米其林星級主廚，經營理念是要帶給顧客飲食及服務上無瑕的享受。

我們專注於提供高單價的美味餐點，主打種類擁有以下：

牛排系列 - 價格：$6000~15000
牛排系列提供許多部位選擇，其中店內招牌為菲力牛排。菲力牛排採用日本A5和牛中的神戶牛，透過3公分厚度的和牛油熟成30天，再由美式碳烤手法過程以桃木煙燻，最後送入烤箱烤出外酥內嫩的極致口感。


餐酒系列 - 價格：$500-45000/每杯(150ml)
餐酒主推82年拉菲，其豐富的層次和深邃口感聞名。結合了黑醋栗、雪松和香料的香氣，擁有完美的平衡感，入口可品嘗出細膩的單寧及悠長的餘韻，是任何珍饈佳餚搭配的不二人選。

我們的營業時間如下：
週一至週日：上午11點至下午22點

備註:
座位稀少可提前來電預約，戰斧牛排須提前3天預訂
"""
```

```python
message = [{"role": "system", "content": system}]
hist = []
def chatbot(text):
  message.append({"role": "user", "content": text})
  res = client.chat.completions.create(
      model="gpt-3.5-turbo",
      messages=message,
      temperature=0
  )
  message.append(res.choices[0].message)
  hist.append([text, res.choices[0].message.content])
  return hist
```

### 效果展示

```python
demo = gr.Interface(
    fn=chatbot,
    inputs="text",
    outputs="chatbot",
    title="餐廳老闆",
    description="請輸入您的問題",
    allow_flagging="never"
)
demo.launch()
```
